# 05. 위젯(Widget) - 02) QLabel    
- URL from : https://wikidocs.net/21939
- `QLabel` 위젯은 텍스트 또는 이미지 라벨을 만들 때 쓴다. 사용자와 어떤 상호작용이 있지는 않다.
- 라벨은 기본적으로 수평 방향으로는 왼쪽, 수직 방향으로는 가운데 정렬이지만 `setAlignment()` 메소드를 통해 조절할 수 있다.

## 예제 스크립트

- 3개의 푸시 버튼을 만들어보자
- 2개의 라벨을 수직 박스에 넣어서 표시해보자

```python
# 필요한 라이브러리 import
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QVBoxLayout
from PyQt5.QtCore import Qt


class MyApp(QWidget) :
    def __init__(self) :
        super().__init__()
        self.initUI()

    def initUI(self) :
        label1 = QLabel('First Label', self)
        label1.setAlignment(Qt.AlignCenter)

        label2 = QLabel('Second Label', self)
        label2.setAlignment(Qt.AlignVCenter)

        font1 = label1.font()
        font1.setPointSize(20)

        font2 = label2.font()
        font2.setFamily('D2Coding')
        font2.setBold(True)

        label1.setFont(font1)
        label2.setFont(font2)

        layout = QVBoxLayout()
        layout.addWidget(label1)
        layout.addWidget(label2)

        self.setLayout(layout)
        self.setWindowTitle('QLabel')
        self.setGeometry(300, 300, 300, 200)
        self.show()

if __name__ == '__main__' :
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_()) 
```

## 예제 스크립트 설명

```python
label1 = QLabel('First Label', self)
label1.setAlignment(Qt.AlignCenter)
```

- `QLabel` 위젯을 하나 만든다. 생성자에는 라벨 텍스트에 부모 위젯을 입력해준다.
- `setAlignment()` 메소드로 라벨의 배치를 설정할 수 있다.
- `Qt.AlignCenter`로 설정해주면 수평, 수직 방향 모두 가운데에 위치하게 된다.

```python
font1 = label1.font()
font1.setPointSize(20)
```

- 라벨에 사용될 폰트를 하나 만든다.
- `setPointSize()` 메소드로 폰트의 크기를 설정해준다.

```python
label2 = QLabel('Second Label', self)
label2.setAlignment(Qt.AlignVCenter)
```

- 두 번재 라벨을 만들고, 이번에는 수직 방향으로만 가운데(`Qt.AlignVCenter`)로 설정해준다.
- 수평 방향으로 가운데로 설정하려면 `Qt.AlignHCenter`를 입력해주면 된다.

```python
font2 = label2.font()
font2.setFamily('D2Coding')
font2.setBold(True)
```

- 두 번째 라벨에 설정할 폰트를 하나 만들고 `setFamily()` 메소드로 폰트의 종류를 설정해준다.
- `setBold(True)`로 폰트를 진하게 설정한다.
- 폰트의 크기를 별도로 설정하지 않았기 때문에 기본값인 13으로 설정된다.
