# 04. 레이아웃(Layout) - 03) 그리드 레이아웃    
- URL from : https://wikidocs.net/21946
- 가장 일반적으로 쓰이는 레이아웃 클래스는 `그리드 레이아웃(grid layout)`이다. 이 레이아웃 클래스는 위젯의 공간을 `행(row)`과 `열(column)`로 구분한다.
- 그리드 레이아웃을 생성하기 위해 `QGridLayout` 클래스를 사용한다.

## 예제 스크립트

- 창의 가운데 아래에 2개의 버튼을 배치시켜보자.
- 2개의 버튼은 창의 크기를 변화시켜도 같은 자리에 위치한다.
- 3개의 라벨, 2개의 라인 에디터, 하나의 텍스트 에디터를 그리드 형태로 배치해보자.

```python
# 필요한 라이브러리 import
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QGridLayout, QLabel, QLineEdit, QTextEdit

class MyApp(QWidget) :
    def __init__(self) :
        super().__init__()
        self.initUI()

    def initUI(self) :
        grid = QGridLayout()
        self.setLayout(grid)

        grid.addWidget(QLabel('Title :'), 0, 0)
        grid.addWidget(QLabel('Author :'), 1, 0)
        grid.addWidget(QLabel('Review :'), 2, 0)

        grid.addWidget(QLineEdit(), 0, 1)
        grid.addWidget(QLineEdit(), 1, 1)
        grid.addWidget(QTextEdit(), 2, 1)
              
        self.setWindowTitle('QGridLayout')
        self.setGeometry(300, 300, 300, 200)
        self.show()

if __name__ == '__main__' :
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_()) 
```

## 예제 스크립트 설명

```python
grid = QGridLayout()
self.setLayout(grid)
```

- `QGridLayout`을 만들고, 앱 창의 레이아웃으로 설정한다.

```python
grid.addWidget(QLabel('Title :'), 0, 0)
grid.addWidget(QLabel('Author :'), 1, 0)
grid.addWidget(QLabel('Review :'), 2, 0)
```

- `addWidget()` 메소드의 첫 번째 위젯은 추가할 위젯, 두, 세번째 위젯은 각각 행과 열 번호를 입력한다.
- 3개의 라벨을 첫 번째 열에 수직으로 배치한다.

```python
grid.addWidget(QTextEdit(), 2, 1)
```

- `QTextEdit()` 위젯은 `QLineEdit()` 위젯과 달리 여러 줄의 텍스트를 수정할 수 있는 위젯이다. 세 번째 행, 두 번째 열에 배치한다.
