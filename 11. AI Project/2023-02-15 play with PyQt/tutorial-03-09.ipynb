## 날짜 표시하기(QDate)
- QDate 클래스는 날짜와 관련된 기능들을 제공한다.

### 현재 날짜 출력하기  
- 우선 QDate 클래스를 이용해서 날짜를 출력해보자
- currentDate() 메소드는 현재 날짜를 반환한다.
- toString() 메소드를 통해 현재 날짜를 문자열로 출력할 수 있다.

````python
from PyQt5.QtCore import QDate

now = QDate.currentDate()
print(now.toString())
````

### 날짜 형식 설정하기  
- toString() 메소드의 format 파라미터를 설정하여 날짜의 형식을 정할 수 있다.
- 'd'는 일(day), 'M'은 달(month), 'y'는 연도(year)를 나타낸다. 각 문자의 갯수에 따라 날짜의 형식이 다르게 출력된다.
- ```Qt.ISODate```, ```Qt.DefaultLocaleLongDate```를 입력하면 ISO 표준 형식 또는 앱의 기본 설정에 맞게 출력할 수 있다.

````python
from PyQt5.QtCore import QDate, Qt

now = QDate.currentDate()
print('d.M.yy 형식은 >>>', now.toString('d.M.yy'))
print('dd.MM.yyyy 형식은 >>>', now.toString('dd.MM.yyyy'))
print('ddd.MMMM.yyyy 형식은 >>>', now.toString('ddd.MMMM.yyyy'))
print('Qt.ISODate 형식은 >>>', now.toString(Qt.ISODate))
print('Qt.DefaultLocaleLongDate 형식은 >>>', now.toString(Qt.DefaultLocaleLongDate))
````

## 시간 표시하기(QTime)
- QTime 클래스를 사용해서 현재 시간을 출력할 수 있다.

### 현재 시간 출력하기
- ```currentTime()```메소드는 현재 시간을 반환한다.
- ```toString()```메소드는 현재 시간을 문자열로 반환한다.

````python
from PyQt5.QtCore import QTime

time = QTime.currentTime()
print(time.toString())
````

### 시간 형식 설정하기
- 'h'는 시간(hour), 'm'은 분(minute), 's'는 초(second) 그리고 'z'는 1/1000초를 나타낸다.
- 또한 날짜랑 마찬가지로 ```Qt.DefaultLocaleLongDate``` 또는 ```Qt.DefaultLocaleShortDate``` 등으로 시간의 형식을 설정할 수 있다.

````python
from PyQt5.QtCore import QTime, Qt

time = QTime.currentTime()
print('h.m.s 형식은 >>>', time.toString('h.m.s'))
print('hh.mm.ss 형식은 >>>', time.toString('hh.mm.ss'))
print('hh.mm.ss.zzz 형식은 >>>', time.toString('hh.mm.ss.zzz'))
print('Qt.DefaultLocaleLongDate 형식은 >>>', time.toString(Qt.DefaultLocaleLongDate))
print('Qt.DefaultLocaleShortDate 형식은 >>>', time.toString(Qt.DefaultLocaleShortDate))
````

## 날짜와 시간 표시하기(QDateTime)
- QDateTime 클래스를 사용해서 현재 날짜와 시간을 함께 출력할 수 있다

### 현재 날짜와 시간 출력하기
- ```currentDateTime()``` 메소드는 현재의 날짜와 시간을 반환한다.
- ```toString()``` 메소드는 날짜와 시간을 문자열 형태로 반환한다.

````python
from PyQt5.QtCore import QDateTime

datetime = QDateTime.currentDateTime()
print(datetime.toString())
````

### 날짜와 시간 형식 설정하기
- 상기 예제에서와 마찬가지로 날짜에 대해 'd', 'M', 'y', 시간에 대해 'h', 'm', 's' 등을 사용해서 날짜와 시간이 표시되는 형식을 설정할 수 있다.
- 또한 ```Qt.DefaultLocaleLongDate``` 또는 ```Qt.DefaultLocaleShortDate```를 입력할 수 있다.

````python
from PyQt5.QtCore import QDateTime, Qt

datetime = QDateTime.currentDateTime()
print('d.M.yy hh:mm:ss 형식 >>>', datetime.toString('d.M.yy hh:mm:ss'))
print('dd.MM.yyyy hh:mm:ss 형식 >>>', datetime.toString('dd.MM.yyyy hh:mm:ss'))
print('Qt.DefaultLocaleLongDate 형식 >>>', datetime.toString(Qt.DefaultLocaleLongDate))
print('Qt.DefaultLocaleShortDate 형식 >>>', datetime.toString(Qt.DefaultLocaleShortDate))
````

````python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow
from PyQt5.QtCore import QDate, Qt

class MyApp(QMainWindow) :
    def __init__(self) :
        super().__init__()
        self.date = QDate.currentDate()
        self.initUI()

    def initUI(self) :
        self.statusBar().showMessage(self.date.toString(Qt.DefaultLocaleLongDate))
        
        self.setWindowTitle('Date')
        self.setGeometry(300, 300, 400, 200)
        self.show()

# 하단 셀의 if __name__ == '__main__'문 안 쓰고 주석 풀고 이거 써도 동작함
# app = QApplication(sys.argv)
# ex = MyApp()
# sys.exit(app.exec_())
````
