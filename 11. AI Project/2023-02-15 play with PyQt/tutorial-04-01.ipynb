# 04. 레이아웃(Layout) - 01) 절대적 배치    
- URL from : https://wikidocs.net/21944
- 절대적 배치(Absolute positioning) 방식은 각 위젯의 위치, 크기를 픽셀 단위로 설정해서 배치한다.
- 절대 배치 방식에는 아래와 같은 제약사항이 있다.
    - 창의 크기를 조절해도 위젯의 크기, 위치가 변하지 않는다.
    - 다양한 플랫폼에서 앱이 다르게 보일 수 있다.
    - 앱의 폰트를 바꾸면 레이아웃이 망가질 수 있다.
    - 레이아웃을 바꾸고 싶다면 완전히 새로 고쳐야 한다.
- 본 예제에서는 2개의 라벨과 2개의 푸시버튼 위젯을 절대적 배치 방식으로 배치해보자

## 예제 스크립트

- 위젯의 위치를 설정하기 위해 `move()` 메소드를 사용하자.
- 라벨(`label1`, `label2`)과 푸시버튼(`btn1`, `btn2`)의 x, y좌표를 설정함으로써 위치를 조절한다.
- 좌표계는 좌측 상단 모서리에서 시작한다. x 좌표는 좌에서 우로 갈수록 커지고, y 좌표는 상에서 하로 갈수록 커진다.

```python
# 필요한 라이브러리 import
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QPushButton

class MyApp(QWidget) :
    def __init__(self) :
        super().__init__()
        self.initUI()

    def initUI(self) :
        label1 = QLabel('Label1', self)
        label1.move(20, 20)
        label2 = QLabel('Label2', self)
        label1.move(20, 60)

        btn1 = QPushButton('Button1', self)
        btn1.move(80, 13)
        btn2 = QPushButton('Button2', self)
        btn1.move(80, 53)
        
        self.setWindowTitle('Absolute Positioning')
        self.setGeometry(300, 300, 400, 200)
        self.show()

if __name__ == '__main__' :
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_()) 
```

## 예제 스크립트 설명
- 라벨을 하나 만들고 x=20, y=20에 위치하도록 옮겨준다.

```python
label1 = QLabel('Label1', self)
label1.move(20, 20)
```

- 푸시 버튼을 하나 만들고, x=80, y=13에 위치하도록 옮겨준다.

```python
btn1 = QPushButton('Button1', self)
btn1.move(80, 13)
```
